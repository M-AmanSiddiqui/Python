

## üéØ Example Data:

Maan lo marks ka ek column hai:

```
Marks = [10, 20, 30, 40, 100]
```

---

## üîπ 1Ô∏è‚É£ Mean ‚Üí ‚ÄúAverage value‚Äù

Formula:
[
\text{Mean} = \frac{\text{Sum of all values}}{\text{Total number of values}}
]

üëâ Calculation:

```
(10 + 20 + 30 + 40 + 100) / 5 = 200 / 5 = 40
```

‚úÖ **Mean = 40**

üí° Matlab: sab numbers ka *average*.

---

## üîπ 2Ô∏è‚É£ Median ‚Üí ‚ÄúMiddle value‚Äù

Steps:

1. Data ko **ascending order** me arrange karo.
   ‚Üí [10, 20, 30, 40, 100]
2. Middle value pick karo (agar odd count hai).

‚úÖ **Median = 30**

üí° Matlab: 30 ke left aur right dono side me equal numbers hain.

---

## üîπ 3Ô∏è‚É£ Mode ‚Üí ‚ÄúMost frequent value‚Äù

üëâ Jo number **sabse zyada baar aaya ho**, wahi mode hota hai.

Example:

```
Marks = [10, 20, 30, 30, 40, 100]
```

‚úÖ **Mode = 30** (kyunki 30 do baar aaya).

---

## üìä Summary Table

| Concept    | Meaning                    | Formula / Rule     | Example Result |
| ---------- | -------------------------- | ------------------ | -------------- |
| **Mean**   | Average of all values      | (Sum √∑ Count)      | 40             |
| **Median** | Middle value (sorted data) | Sort ‚Üí Pick middle | 30             |
| **Mode**   | Most frequent value        | Highest frequency  | 30             |

---

## üí° When to Use Which?

| Use Case                                    | Best Choice | Reason                         |
| ------------------------------------------- | ----------- | ------------------------------ |
| Data balanced & clean                       | **Mean**    | Shows overall average          |
| Data has outliers (e.g., 100 is very high)  | **Median**  | Not affected by extreme values |
| Categorical data (like Gender, City, Grade) | **Mode**    | Shows most common value        |

---

## üìà Visualization (Conceptually)

```
10   20   30   40   100
 |----|----|----|----|
        ‚Üë   ‚Üë       ‚Üë
       Mode Median  Mean (pulled by 100)
```

üí¨ Mean shift ho jata hai jab data me **bohot bada ya chhota number (outlier)** ho.

---

## ‚úÖ Simple Summary:

| Term       | Simple Definition   | Example                  |
| ---------- | ------------------- | ------------------------ |
| **Mean**   | Average             | (10+20+30+40+100)/5 = 40 |
| **Median** | Middle value        | 30                       |
| **Mode**   | Most repeated value | 30                       |
